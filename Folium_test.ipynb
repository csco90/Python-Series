## Zooming in.
### Do you Recognize this area...?

```python

m=folium.Map(
    location=[25.796821, -80.248675],
    tiles='Stamen Toner',
    zoom_start=12
)

folium.CircleMarker(
    location=[25.794496, -80.282314],
    radius=20,
    popup='Come to Miami. Don\'t be MIA!',
    color='orange',
    fill=True,
    fill_color='orange'
).add_to(m)

folium.CircleMarker(
    location=[25.781094, -80.193298],
    radius=20,
    popup='Miami Downtown',
    color='red',
    fill=True,
    fill_color='red'
).add_to(m)

folium.CircleMarker(
    location=[25.764518, -80.192244],
    radius=20,
    popup='Brickell',
    color='green',
    fill=True,
    fill_color='green'
).add_to(m)

folium.CircleMarker(
    location=[25.861128, -80.297739],
    radius=40,
    popup='Do not forget Hialeah',
    color='blue',
    fill=True,
    fill_color='blue'
).add_to(m)

m
```
